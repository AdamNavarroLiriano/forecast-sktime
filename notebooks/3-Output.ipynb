```python
import pandas as pd
import numpy as np 
from sktime.utils.plotting import plot_series
```

```python
train_data = pd.read_csv('../data/train.csv', index_col=0)
test_data = pd.read_csv('../data/test.csv', index_col=0)

train_data.index = pd.period_range('2015.01', freq='M', periods=train_data.shape[0])
test_data.index = pd.period_range('2021.03', freq='M', periods=test_data.shape[0])
```

```python
plot_series(train_data['y'], test_data['y'], labels=['actual', 'predicted'], title='Predictions')
```
